```python
import pandas as pd
```

```python
# reading .json file using read_json(file_path) method

df = pd.read_json('input.json')
print('Dataframe Loaded:\n')
print(df)
```

```python
# converting dataframe to .json or .csv using 

# to_csv(file_path, index)
df.to_csv('output_csv.csv', index=False)

# to_json(file_path, orient, indent)
# by default, orient is set as index (i.e index will be the key)
# if orient == records, then the key will be the column names

df.to_json('output_json.json', orient='records', indent=4)
```

```python
# reading table from .html page (online webpage) using .read_html(url)

url = "https://www.fdic.gov/resources/resolutions/bank-failures/failed-bank-list/"
dfs = pd.read_html(url)          # returns a list of all tables present in the webpage

df = dfs[0]         # use 1st table
print(df.head(10))

df.to_json('output_json.json', indent=4, orient='records')
```

```python
# similar to .csv file, we can also read .xlsx file
# Syntax :- df = pd.read_excel(file_path, sheet_name)
```

```python
# pickle file - it is a python object which is obtained from serialising a python object into byte stream

df = pd.read_csv("sales_data.csv")
print(df)
print()

df.to_pickle('output_pickle.pkl')

file = open("output_pickle.pkl", 'rb')
print("Picke File Written:")
print([x for x in file])
print()
file.close()

print("Pickle File Read:")
pkl_to_df = pd.read_pickle('output_pickle.pkl')
print(pkl_to_df)
```
