```python
# file operations
# Syntax :- file = open("file_path", 'open-mode')
# 'r' - read mode           | 'w' - write mode (overwriting)
# 'a' - append mode
# 'rb' - read byte mode     | 'wb' - write byte mode
# 'ab' - append byte mode

# additonal file-open modes
# 'a+' - open file in R/A mode (if file doesn't exist then create it)
# 'w+' - open file in R/W mode (if file doesn't exist then create it)
# 'r+' - open file in R/W mode (if file doesn't exist then error)

# complete mode list :-
# a - append        b - binary(byte-wise)
# w - write         t - text
# r - read          + - Read and Write
```

```python
file = open("example.txt", 'r')

# read() function returns the complete content of file in string format
content = file.read()               
print(content)
file.close()                    # must close before reading/opening again
print()

file = open("example.txt", 'r')

# reading line by line
for line in file:
    line = line.strip()         # .strip() removes any newline character
    print(line)
file.close()
```

```python
# opening a file in write mode (write with overwriting)
file = open('write_example.txt', 'w')

file.write('This is an example of write operation')
file.write('\n')
file.write('THis is ML Course with python')

file.close()
```

```python
# opening a file in append mode (write without overwriting)
file = open('write_example.txt', 'a')

file.write('\n')
file.write('Hello Word')
file.write('\n')
file.write('This is an example of appended content')

file.close()
```

```python
# opening a file in read-write mode
file = open("read_write_example.txt", 'w+')
file.write("Hello World\n")
file.write("Welcome to ML Course with Python\n")

# move the cursor to the beginning of the file to read (now the cursor is at end)
file.seek(0)
content = file.read()
print(content)
```

```python
# read a file and count no of lines, words and characters
file = open("example.txt", 'r')
content = file.read()
lines = content.split('\n')

total_lines = len(lines)
total_words = 0
total_characters = 0

for line in lines:
    total_characters += len(line)
    words = line.split()
    total_words += len(words)

print("Total Lines = ", total_lines)
print("Total Words = ", total_words)
print("Total Characters = ", total_characters)

file.close()
```
