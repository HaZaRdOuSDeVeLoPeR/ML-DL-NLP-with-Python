```python
import random

# binary file operations
# writing binary file (byte-by-byte)
file = open('binary_file.bin', 'wb')

file.write(bytes([random.randint(0,255)]))
file.write(bytes([random.randint(0,255)]))
file.write(bytes([random.randint(0,255)]))

file.close()
```

```python
# reading binary file
file = open('binary_file.bin', 'rb')

content = file.read()
print(list(content))
file.close()
```
