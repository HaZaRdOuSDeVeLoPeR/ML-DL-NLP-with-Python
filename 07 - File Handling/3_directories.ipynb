```python
import os

# make new folder in current directory
new_dir = 'test_dir'
os.mkdir(new_dir)

# joining the paths, joins using '\' in windows and '/' in linux
new_path = os.path.join(os.getcwd(), new_dir)

print(f"{new_path} created")
```

```python
# listing directory files
# Syntax :- os.listdir(path)
import os

items = os.listdir(os.getcwd())
print(items)
```

```python
# checking whether a path is a file or a directory
import os

path = "test_dir"

if os.path.isfile(path):                # isfile(path) : checking for file
    print(f"{path} is a file")
elif os.path.isdir(path):               # isdir(path) : checking for directory
    print(f"{path} is a directory")
else:
    print(f"{path} is neither a file or a directory")
```
