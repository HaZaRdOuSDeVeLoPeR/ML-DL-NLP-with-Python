```python
# Probability :-
# Probability is all about determining the likelihood of an event occurring.


# Addition Rule for Mutually / Non-Mutually Exclusive Events :-
#   |
#   |-----> Mutually Exclusive Events :- If two events can not occur at same time
#   |       Eg :- P(Head) & P(Tail) in coin toss  |   P(1) & P(2) in in die roll
#   |
#   |       In such events :- P(A or B) = P(A) + P(B)
#   |
#   |-----> Non-Mutually Exclusive Events :-  If two events may occur at same time
#           Eg :- P(King) & P(Red) in a pack of cards
#    
#           In such events :- P(A or B) = P(A) + P(B) - P(A and B)


# Multiplication Rule for Dependent / Independent Events :-
#   |
#   |-----> Independent Events :- If the outcome of one event can't affect the other event
#   |       Eg :- P(Tail) after getting (Head) in a Coin Toss
#   |
#   |       In such events :- P(A and B) = P(A) * P(B)
#   |
#   |-----> Dependent Events :-  If the outcome of one event affects the other event
#           Eg :- P(Red Card) after drawing (King)
#    
#           In such events :- P(A and B) = P(A|B) * P(B) = P(B|A) * P(A)    {Bayes' Theorem}
```
