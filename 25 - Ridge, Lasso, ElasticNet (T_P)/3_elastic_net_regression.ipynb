```python
# ElasticNet Regression

# It is combination of Both Ridge and Lasso Regression

# Can do both simultaneously :-
#       - Avoid Overfitting (Control size of coefficients)
#       - Feature Selection (By Shrinking the coefficients of useless features to 0)

# Advantages of Using Elastic Net :-
#       - Groups correlated features together
#       - Selects some features automatically
#       - Shrinks coefficients to prevent overfitting
```

```python
# Usual Cost Function :- 
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²]
# 
#                                        (L1 Regularⁿ)  (L2 Regularⁿ)  
# Modified Cost Function :-                     |           |
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²] + λ₁.∑|βⱼ| + λ₂.∑(βⱼ²) 
# 
#               Here, λ₁.∑|βⱼ| is the penalty term for L1 Regularization (Lasso Regression)
#                     λ₂.∑(βⱼ²) is the penalty term for L2 Regularization (Ridge Regression)
#                     λ₁ and λ₂ are hyperparameters for L1 and L2 Regularizations respectively
```
