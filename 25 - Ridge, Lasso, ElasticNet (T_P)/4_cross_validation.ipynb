```python
# Cross Validation
# It is a model evaluation technique used to check how well a model generalizes unseen data

# Cross validation helps in identifying overfitting in Model by repeatedly testing on different portions of data
```

```python
# How Cross Validation is applied ?

# Split "Training Dataset" into multiple parts
# Train on some part while Testing on the remaining Part.
```

```python
# Types of Cross-Validation :-

# 1. Hold-Out Validation [Most Simple]
# 
#   - Split Dataset once into two parts [Training + Test] (lets say 70-30 split)
#   - Train on Training Dataset. No Interaction with Test Data while Training Process
#   - After Model is completely traine, Test the Model performance using Test Dataset


# 2. k-Fold Cross Validation [Most Common]
# 
#   - Split Dataset into K equal folds
#   - For each fold :
#       Train on remaining K-1 folds and test on this fold
#   - Repeat k Times
# 
#   - Final Performance Score is the average of all k Tests


# 3. Stratified k-Fold Cross Validation [Most Common and Fair]
# 
#   - Split Dataset into K equal folds such that class proportions is same in each fold
#   - Mostly used in Classification (Equal nos. of Zeroes and Ones)
#   - Prents Class-Imbalance Errors


# 4. Leave-One-Out Cross Validation [Most Extreme]
# 
#   - Similar to k-Fold but here, k = no of observation (n)
#   - For each n datapoints :
#       Train on n-1 datapoints and test on this datapoint
#   - Repeat n Times
# 
#   - Final Performance Score is the average of all n Tests


# 5. Leave-P-Out Cross Validation [Somewhat less extreme]
# 
#   - Similar to Leave One Out but here, we split into "p" and "n-p" datapoints
#   - For each p datapoints :
#       Train on "n-p" datapoints and test on "p" datapoints
#   - Repeat n-p Times
# 
#   - Final Performance Score is the average of all n-p Tests


# 6. Time Series Cross Validation [Rolling/Expanding Window]
# 
#   - Training Set Grows over time, Test set shrinks
```
