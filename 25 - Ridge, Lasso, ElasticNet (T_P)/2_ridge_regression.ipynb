```python
# Ridge Regression

# This technique is also known as L2 Regularization.

# For special cases when :-
#   i) Multiple Features are Correlated (Multicollinearity)
#  ii) Dataset is Small
# iii) Matrix XᵀX is nearly singular 
# 
# In such conditions βᵢ becomes unstable.

# Ridge Regression prevents from "overfitting" by shrinking the coefficients(βᵢ) towards zero
```

```python
from IPython.display import Image, display
display(Image(filename = "./assets/ridge1.png"))
display(Image(filename = "./assets/ridge2.png"))
```

```python
# It works by adding a penalty term to the cost function :-

# Usual Cost Function :- 
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²]
# 
# Modified Cost Function :-
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²] + λ.∑[βⱼ²]
#                                             Here, λ.∑[βⱼ²] is the penalty term

# The strength of Regularization can be altered by altering value of λ (hyperparameter)


# Overfitting happens when:
# - Model learns noise and small fluctuations.
# - Coefficients become huge to perfectly fit all training points.
# - The fitted line wiggles too much → poor generalization.

# Ultimate Goal :-
# Reduce Overfitting by controlling Size(Magnitude) of coefficients
# In Simple Words, Find a Fine Balance between Fitting the Data and keeping coefficients small
#                                                                     [Avoiding Overfitting]

# Note :- 
# Increasing λ shrinks the coefficients towards 0 but it will never reach 0
```
