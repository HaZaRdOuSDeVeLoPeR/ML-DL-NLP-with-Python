```python
# Lasso Regression

# This technique is also known as L1 Regularization.
# Mainly used for "Feature Selection"

# The coefficients of unnecessary features actually shrinks to zero.
# Lasso adds the penalty for each coefficient equivalent to its magnitude |βᵢ|

# Due to this property, it is used for feature selection 
```

```python
from IPython.display import Image, display
display(Image(filename = "./assets/lasso1.png"))
display(Image(filename = "./assets/lasso2.png"))
```

```python
# Usual Cost Function :- 
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²]
# 
# Modified Cost Function :-
#               J = (1/n) * ∑[(yᵢ - yᵢᵖ)²] + λ.∑|βⱼ|
#                                             Here, λ.∑|βⱼ| is the penalty term

# The strength of Regularization can be altered by altering value of λ (hyperparameter)

# Ultimate Goal :-
# Shrink the coefficients of useless features heavily to the point where it can be discarded
# Shrink the coefficients of useful features mildly keeping them relevant
# Coefficients of useless features actually shrinks to 0

# Note :- 
# Small λ shows mild regularization
# Large λ shows rapid shrinkage of coefficients, many become zero
# Very large λ causes underfitting in Model
```

```python
# When Should We use Lasso Regression ?

# - When there are too many input features and we don't know which features to select
# - In short for feature selection
```
