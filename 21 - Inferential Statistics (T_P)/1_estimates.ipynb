```python
# Estimates

# Estimate is a specified observed numerical value 
# used to estimate an unknown population parameter

# Types of Estimates
#   |
#   |------> Point Estimate :- Single Numerical value used to estimate an unknown population parameter
#   |           Eg :- Using sample mean to estimate population mean
#   |
#   |------> Interval Estimate :- Range of values is used to estimate unknown population parameter
#               Eg :- Using the interval [μ-k.σ, μ+k.σ] to estimate population mean
#               - The estimated interval is known as confidence interval        
```
