```python
# Hypothesis Testing is a technique used in inferential statistics 
# which is used to draw some conclusions about the population 
# using sampled data

# Hypothesis Testing Mechanism :-
# 
# Step 1 :- Define Null Hypothesis (H0) : Assumption which is considered to be true
# Step 2 :- Alternate Hypothesis (H1) : Opposite of Null Hypothesis
#           Note :- H0 must contain equality interval (i.e =, >= or <=)
# 
# Step 3 :- Define all the population and sample parameters
# Step 4 :- Perform various tests and accept/reject the null hypothesis
```

```python
# P Value

# P value is a number calculated by performing a statistical test.
# It describes how likely it is to find a particular set of observations if H0 were true
# It is used to decide whether to accept/reject Null Hypothesis


# Significance Level

# Significance level is the probability of rejecting a true null hypothesis while performing a hypothesis test.
# It represents the maximum risk of making a Type I error.
```
