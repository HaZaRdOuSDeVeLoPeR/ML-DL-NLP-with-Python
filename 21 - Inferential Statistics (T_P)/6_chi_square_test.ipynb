```python
# Chi-Square Test

# It is a non-parametric test which is performed on catgorical [ordinal, nominal] data
# Used in Testing whether thr given data fits any known theoritical distribution
#       - How well a data fits in a theoritical distribution/data (Goodness of Fit Test)

# Steps :-
# Step 1 :- Compute Observed(Given) Frequency and Expected(Theoritical) Frequencies 
#           for each datapoints for the required theoritical distribution/data 

# Step 2 :- Compute the 2 value for the given distribution from Oi and Ei

# Step 3 :- Always Double Tailed Test is Performed
#               - H0 : Data Fits the Distribution
#               - H1 : Data Doesn't Fits the Distribution

# Step 4 :- Compute 2(alpha, df) from the given confidence level (LOS) and degree of freedom (df = k - 1 - p)
#           - Determine the confidence level area and boundaries using 2 table

# Step 5 :- If 2 value lies within the confidence region {region constituting 2 = 0}
#           - then accept H0
#           - else reject

# Example :-

# Given Population Distribution in 2010
# Group     |   <50kg   |   50-75kg     |   >75kg
# Frequency |    100    |     150       |    250

# Given Population Distribution in 2020
# Group     |   <50kg   |   50-75kg     |   >75kg
# Frequency |    140    |     160       |    200

# Test whether the population distribution has changed over the 10 years ?

# Step 1 :-
# Group             |   <50kg   |   50-75kg     |   >75kg
# Obs Frequency     |    140    |     160       |    200
# Theo Frequency    |    100    |     150       |    250

# Step 2 :-
# 2 value = 26.66

# Step 3 :-
# H0 :- Population fits the earlier distribution (population dist has not changed)
# H1 :- Population doesn't fit the earlier distribution (population dist has changed) 

# Step 4 :-
# df = 3 - 0 - 1 = 2
# 2(0.05, 2) = 5.991

# Step 5 :-
# 2 value > |2(0.05, 2)|      Hence Null Hypothesis is Rejected.
```
