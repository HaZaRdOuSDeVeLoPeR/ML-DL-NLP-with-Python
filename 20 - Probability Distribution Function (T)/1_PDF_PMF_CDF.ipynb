```python
# Probability Distribution Function

# PDF describe how the probabilities are distributed over the values of a random variable.
# It is a histogram with bin size tending to 0

# Types of Probability Distribution Functions
#   |
#   |------> "Probability Mass Function" (Used for discrete Random Variables)
#   |           Eg :- Rolling a fair dice
#   |
#   |------> "Probability Density Function" (USed for continuous Random Variables)


# Properties of PDF
# 
# 1. Function should be non-negative for all value of X, f(X) >= 0
# 2. Total Area under the PDF curve is 1
```

```python
# Cumulative Density Function
# Plot the area under the curve for PDF to get CDF
# It is the integration of probability density function

# We can also say that Probability Density Function is the Gradient of Cumulative Density Function

from matplotlib import pyplot as plotter

X = [x for x in range(1,7)]
P_X = [1/6 for x in range(1,7)]
F_X = [x*1/6 for x in range(1,7)]


plotter.figure(figsize=(8,4))
plotter.subplot(1,2,1)
plotter.xlabel('X')
plotter.ylabel('P(X)')
plotter.title('Probability Mass Function')
plotter.bar(X, P_X, color = ['red', 'blue', 'green', 'yellow', 'pink', 'orange'])

plotter.subplot(1,2,2)
plotter.xlabel('X')
plotter.ylabel('F(X)')
plotter.title('Cumulative Density Function')
plotter.bar(X, F_X, color = ['red', 'blue', 'green', 'yellow', 'pink', 'orange'])
```
