```python
# Probability Distribution Functions follows different kinds of distributions.
# Different distribution have different properties which can be used to make various assumptions related to data.

# There are some common distributions which are often found in nature :-
#
# 1. Bernoulli's Distribution - Outcomes are binary (0/1)
# 
# 2. Binomial Distribution - Generalized version of Bernoulli's Distribution. Outcomes are discrete
# 
# 3. Poisson Distribution - Generalized version of Binomial Distribution when no of experiments is very large
# 
# 4. Normal/Gaussian Distribution - Outcomes are continuous where mean value is most probable
#
# 5. Uniform Distribution - When each event has equal outcome
# 
# 6. Log Normal Distribution - When log of random variable follows normal distribution
# 
# 7. Power Law Distribution
# 
# 8. Pareto Distribution
```

```python
# Bernoulli's Distribution  [Discrete]

# It is the simplest discrete probability distribution.
# Random variable has exactly two possible (binary) outcomes:
#       - success (with probability p)
#       - failure (with probability 1-p)

# Eg :- Tossing a Coin, Predicting Rain

# Parameters used :- 
#   p - Probability of Success (ranges between 0 and 1)
#   q - Probability of failure (ranges between 0 and 1)     q = 1 - p
#   x - outcome of the event âˆˆ {0, 1}

# Mean = p
# Variance = p.q


def Bernoulli_PMF(p, x):
    return p**x * (1-p)**(1-x)

print("Experiement of Tossing a unfair coin")
print("P(Head: 1) = p = 0.6 {Success}", "\nP(Tail: 0) = q = 0.4 {Failure}")
print()
print("P(X = 0):", Bernoulli_PMF(0.6, 0))
print("P(X = 1):", Bernoulli_PMF(0.6, 1))
```

```python
# Binomial Distribution     [Discrete]

# Generalized version of Bernoulli's Distribution where we perform multiple trials.
# Each trial has probability of success "p" and a probability of failure "q" (1-p)
# Outcomes of random variables is binary {0, 1}
# Denoted by B(n, p, k)

# Parameters used :-
#   p, q same as bernoulli's distribution
#   n = no. of trials performed
#   k = no. of successes    k âˆˆ [0,n]

# Mean = n.p
# Variance = n.p.q

# Example
print("Inspecting 20 items in a factory where each item has 10% chance of being defective")
print("Compute the chance of finding k defective items for k âˆˆ [0,20]")

import math
from matplotlib import pyplot as plotter
def Binomial_PMF(n, p, k):
    return (math.comb(n, k)) * (p**k) * ((1-p)**(n-k))

p = 0.4
n = 20
success = [x for x in range(0,n+1)]
P_success = [Binomial_PMF(n, p, k) for k in success]
print()
print([f"{i}: {round(P_success[i],4)}" for i in range(0, 21)])

plotter.title('Binomial Distribution p = 0.4, n = 20')
plotter.xlabel('no of success')
plotter.ylabel('P(Success)')
plotter.xticks([x for x in range(0, n+1, 1)])
plotter.plot(success, P_success, marker='o')
```

```python
# Poisson Distribution      [Discrete]

# Generalized Version of Binomial distribution where 
#                   - if events occur (repeat) in a fixed interval of time
#                   - constant rate is denoted by Î»(events/unit_time)
# Denoted by :- P(Î», k)

# Mean = Variance = Î»*t     ,where t = time interval

# Example :-
print("Find the probability that k persons visit the hospital in next hour")
print("Given Î» = 5/hr, Î» = 10/hr, Î» = 15/hr")
print("For k âˆˆ [0, 25]")

import math
from matplotlib import pyplot as plotter
def Poisson_PMF(lamda, k):
    return ((math.e**(-lamda)) * (lamda**k)) / math.factorial(k)

n = 25
visits = [x for x in range(0, n+1)]
P_visits5 = [Poisson_PMF(5, k) for k in range(0, n+1)]
P_visits10 = [Poisson_PMF(10, k) for k in range(0, n+1)]
P_visits15 = [Poisson_PMF(15, k) for k in range(0, n+1)]

plotter.figure(figsize=(15,5))
plotter.subplot(1,3,1)
plotter.title('for Î» = 5')
plotter.xlabel('visits')
plotter.ylabel('P (visits)')
plotter.xticks([x for x in range(0, n+1, 2)])
plotter.plot(visits, P_visits5, marker='o')

plotter.subplot(1,3,2)
plotter.title('for Î» = 10')
plotter.xlabel('visits')
plotter.ylabel('P (visits)')
plotter.xticks([x for x in range(0, n+1, 2)])
plotter.plot(visits, P_visits10, marker='o')

plotter.subplot(1,3,3)
plotter.title('for Î» = 15')
plotter.xlabel('visits')
plotter.ylabel('P (visits)')
plotter.xticks([x for x in range(0, n+1, 2)])
plotter.plot(visits, P_visits15, marker='o')

# Observations :-
# Peak shifts towards Î»
```

```python
# Normal Distribution    [Continuous]

# Distribution is Bell-curve Shaped (peak at Î¼)
# Distribution is symmetric about mean and spread increases as variance increases
# Used for continuous random Variable
# Denoted by N(Î¼, Ïƒ, x)

# Parameters used :-
# z-score(x) = (x-Î¼)/Ïƒ      [Magnitude of z-score denotes the deviation of "x" w.r.t mean]

# Empirical Rule (68, 95.5, 99.7) of Normal Distribution :-
# 
# 1. in range of [Î¼ - Ïƒ, Î¼ + Ïƒ]   lies 68% of data points
# 2. in range of [Î¼ - 2Ïƒ, Î¼ + 2Ïƒ] lies 95% of data points
# 3. in range of [Î¼ - 3Ïƒ, Î¼ + 3Ïƒ] lies 99.7% of data points

# Mean = Î¼
# Variance = Ïƒ^2

import math
from matplotlib import pyplot as plotter

def Normal_PDF(mu, sigma, x):
    z = (x - mu)/sigma
    return (1/(sigma*math.sqrt(2*math.pi))) * math.e**(-(z**2)/2)

X = [x*0.1 for x in range(0, 1000)]
N_X_50_2 = [Normal_PDF(50, 2, x) for x in X]      # Î¼ = 50, Ïƒ = 2         
N_X_50_5 = [Normal_PDF(50, 5, x) for x in X]      # Î¼ = 50, Ïƒ = 5 
N_X_50_8 = [Normal_PDF(50, 8, x) for x in X]      # Î¼ = 50, Ïƒ = 8 
N_X_50_10 = [Normal_PDF(50, 10, x) for x in X]     # Î¼ = 50, Ïƒ = 10
N_X_50_20 = [Normal_PDF(50, 20, x) for x in X]     # Î¼ = 50, Ïƒ = 20

plotter.xlabel('x')
plotter.ylabel('N(x)')
plotter.title('Normal Distribution')
plotter.plot(X, N_X_50_2, label = 'Î¼ = 50, Ïƒ = 2')
plotter.plot(X, N_X_50_5, label = 'Î¼ = 50, Ïƒ = 5')
plotter.plot(X, N_X_50_8, label = 'Î¼ = 50, Ïƒ = 8')
plotter.plot(X, N_X_50_10, label = 'Î¼ = 50, Ïƒ = 10')
plotter.plot(X, N_X_50_20, label = 'Î¼ = 50, Ïƒ = 20')
plotter.legend()
plotter.show()
```

```python
# Standard Normal Distribution

# Every normal distribution can be transformed into standard normal distribution.
# For every datapoint, apply a transformation :-
#       z-score(x) = (x-Î¼)/Ïƒ                    [i.e replace x by its z-score]

# Standard normal distribution has Î¼ = 0 and Ïƒ = 1

# Practical Importance
# 1. Now we can comment about any datapoint using empirical formula easily (since Î¼ = 0 and Ïƒ = 1)
# 2. While training a ML Model, every features are measured in differnt units
#       --> we use "standardization" to normalize the datapoints for each feature

# Example :-
import numpy as np

un_normalized_data = np.array([2,4,6,7,9,11,17,25,30,32,37,42,45])
print("Original Data:", un_normalized_data.tolist())
print("Mean:", un_normalized_data.mean())
print("Standard Deviation:", un_normalized_data.std())
print()

normalized_data = np.array([round(((x-un_normalized_data.mean())/un_normalized_data.std()),3) for x in un_normalized_data])
print("Normalized Data:", normalized_data.tolist())
print("Mean:", round(normalized_data.mean(),2))
print("Standard Deviation:", round(normalized_data.std(),2))
```

```python
# Uniform Distribution      [Both Continuous and Discrete]

# Every event in the range [a, b] has the same probability

# Mean = (a + b)/2
# Variance = {(b - a)^2}/12

# Example
print("Number of candies sold daily at a shop is uniformly distributed with a maximum of 40 candies and a minimum of 10")
print("Find the probability for the daily sales to fall between 15 and 30")
print()

def Uniform_PDF(a, b):
    return 1/(b-a)

a = 10; b = 40
sold = [x for x in range(a, b+1)]
P_sold = [Uniform_PDF(a, b) for x in sold]

P_16_29 = [x for x in P_sold[sold.index(16):sold.index(30)]]
sum = 0
for x in P_16_29:
    sum += x

print("Required Probability:", sum)
from matplotlib import pyplot as plotter
plotter.title('Uniform Distribution')
plotter.xlabel('sold')
plotter.ylabel('P (sold)')
plotter.plot(sold, P_sold)
```

```python
# Log Normal Distribution       [Continuous]
# Used where the natural logarithm of the random variable follows normal distribution.

# If X is log-normally distributed then Y = ln(X) will be normally distributed
#                           conversely
# If Y is normally distributed then e^Y will be log-normally distributed

# Eg :- Wealth Distribution among the people in the world

import math
from matplotlib import pyplot as plotter
import numpy as np

normalZ = [x*0.001 for x in range(-2500, 2500)]
normal_dist = [Normal_PDF(0, 1, x) for x in normalZ]

def Log_Normal_PDF(mu, sigma, x):
    return (1/(x*sigma*math.sqrt(2*math.pi))) * math.e**-((math.log(x, math.e)-mu)**2/(2*sigma**2))

log_normalZ = [math.e**x for x in normalZ]
log_normal_dist = [Log_Normal_PDF(0, 1, x) for x in log_normalZ]

plotter.figure(figsize=(10,5))
plotter.subplot(1,2,1)
plotter.title('Standard Normal Distribution')
plotter.plot(normalZ, normal_dist)

plotter.subplot(1,2,2)
plotter.title('Standard Log Normal Distribution')
plotter.plot(log_normalZ, log_normal_dist)

# Observation :-
# The peak is skewed (towards left or right)
# The distribution has a heavy tail
```
