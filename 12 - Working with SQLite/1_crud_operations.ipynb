```python
# SQL is a standard language for managing and manipulating relational databases.
#                                    whereas
# SQLite is a self-contained, serverless and zero configuration database engine 
# mostly used in embedded database systems.

# in python3, sqlite3 comes preinstalled

# Datatypes in SQLite3 :-
# Integer, Real, Text, BLOB only
```

```python
import sqlite3 as sql
```

```python
# connect to an SQLite database (a local .db file)
# creates a database / links to a pre-existing database
database = sql.connect('example.db')
print(database)

# create a cursor for the database
cursor = database.cursor()
print(cursor)
```

```python
# Creating a table
create_table_query = """
    create table if not exists Employees(
        Emp_ID Integer Primary Key,
        Name Text Not Null,
        Age Integer,
        Department Text
    )
"""
cursor.execute(create_table_query)
```

```python
# Insertion in table
id = int(input("Enter Employee ID: "))
name = input("Enter Employee Name: ")
age = int(input("Enter the Employee's Age: "))
dept = input("Enter the Department: ")

insert_query = f"""
    Insert Into Employees(Emp_ID, Name, Age, Department)
    Values({id}, '{name}', {age}, '{dept}')
"""

cursor.execute(insert_query)
```

```python
# Running some query
select_all = """
    select * from Employees
"""
cursor.execute(select_all)

records = cursor.fetchall()
for record in records:
    print(record)
```

```python
# Updating the table
update_query = """
    update Employees
    set Department = "Machine Learning"
    where Name = "Aditya Vimal"
"""

cursor.execute(update_query)

cursor.execute(select_all)
records = cursor.fetchall()
for record in records:
    print(record)
```

```python
database.commit()       # commits the changes the local .db file

# Deleting the records
delete_query = """
    delete from Employees
    where Name = 'Aditya Vimal'
"""

cursor.execute(delete_query)

cursor.execute(select_all)
records = cursor.fetchall()
for record in records:
    print(record)
```

```python
database.rollback()     # rolls back to last commit (or null if no commits made yet)

cursor.execute(select_all)
records = cursor.fetchall()
for record in records:
    print(record)

database.commit()       # commits the changes the local .db file
```

```python
database.close()            # closes the connection
```
