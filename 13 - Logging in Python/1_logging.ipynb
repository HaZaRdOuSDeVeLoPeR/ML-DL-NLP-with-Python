```python
# logging is a way to track events, errors and operational information.
# Python has built-in logging module for emitting log messages from Python Programs
```

```python
import logging
```

```python
# configuring the basic logging settings
logging.basicConfig(
    filename = 'app.log',           # save logging messages in app.log file
    filemode = 'w+',                # open mode for log file
    level = logging.DEBUG,
    format = '%(asctime)s-%(name)s-%(levelname)s-%(message)s',      # format for logging message
    datefmt= '%Y-%m-%d %H:%M:%S'                                    # format for asctime
)

# logging levels (sorted on basis of priority - low to high)

logging.debug("This is a debug message")        # detailed info, for diagnosing problems
logging.info("This is a info message")          # confirmation for everything working as expected
logging.warning("This is a warning message")    # some indication of unexpected behaviour, problem may arise in near future
logging.error("This is a error message")        # due to a serious problem, program is not able to perform some function
logging.critical("This is a critical message")  # very serious error, indicating program may itself be halted
```
