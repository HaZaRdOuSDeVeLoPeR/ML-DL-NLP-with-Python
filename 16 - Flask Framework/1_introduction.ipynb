```python
# Flask is a complete web framework based on python used in creating end-to-end Web Application
# Two Important Components in Flask Framework :-

# 1. WSGI (Web Server Gateway Interface)
# Architecture :-

# [Using HTTP/TCP Protocol]                     [Using WSGI Protocol]
#                 (Request)    |------------|    (Request)   |-----------|
#                ------------->|            |--------------->|           |   {Deployed
# |User Requests|              |    Web     |                |    Web    |      on
#                <-------------|   Server   |<---------------|    App    |  Web Server}
#                 (Response)   |------------|    (Response)  |-----------|
#                                   \/                            \/
#                           [AWS, AZURE, APACHE]            [Built using Flask]

# 2. Jinja 2 Template Engine :- 
# It is a web template engine.
#       --> Combines a web template (webpage) with a data source to create "dynamic webpages"
#       --> data sources can be SQL, CSV, ML Model, MongoDB etc

# 3. Flask provides multiple objects for abstraction
#       --> request object - to access multiple objects from user request
#               - request.form (access form data) 
#               - request.args (access url parameters)
#               - request.method (access the request method from user)
# 
#       --> session object - to assign cookies to the browser
#               - one unique object for each web client 
#               - can be used to uniquely identify a session between server and client
#               - must set a secret key for calculation of cookies for each flask instance using session object

# 4. all the jinja2 templates must be in "templates" folder otherwise must assign while making the flask app using "template_folder" parameter

# 5. all the static files must be in "static" folder otherwise must assign while making the flask app using "static_folder" parameter
```
