```python
# using range() function to generate a sequence of numbers
#   Syntax :- range(start, stop, step)      here, stop is not included

# use for loop to traverse in range 
# loop variable is automatically incremented
#   Syntax :- for var_name in Object_Name:
          
for i in range(1, 5):                    # 5 is not included
    print(i)

ten_twenty_skip2 = range(10, 20, 2)       # 20 is not included
for i in ten_twenty_skip2:
    print(i)

#######################################################################

# use while loop if count of interations is known beforhand
# loop variable must be incremented manually
#   Syntax :- while condition:
count = 0
while count < 5:
    print("Inside while loop")
    count = count + 1
```

```python
# loop control statement
#   1. break - exits the loop prematurely if some condition becomes true
#   2. continue - restart the loop (and increment the loop variable incase of for loop)

for i in range(1, 10):
    print(i)
    if(i == 4): break           # break the loop at 4 (before reaching 10)

for i in range(20, 30):
    if(i % 2 == 0): continue    # skips all the even numbers
    print(i)
```

```python
# iterating over string
mystr = "Aditya Vimal"
for c in mystr:
    print(c, end=", ")
```
