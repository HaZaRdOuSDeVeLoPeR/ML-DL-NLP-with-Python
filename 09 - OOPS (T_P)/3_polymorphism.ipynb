```python
# Polymorphism is a core concept of OOP which allows objects child class to be treated as objects of common superclass.
# Usually achieved through method overriding and interfaces

# function overriding is used in terms of OOP and function overloading in terms of FOP
```

```python
# Base Class
class Animal:
    def __init__(self): pass
    def speak(self):
        return "Sound of an Animal"
    
# Derived Class 1
class Dog(Animal):
    def __init(self): pass
    def speak(self):        # overriding speak() function of parent class
        return "Woof !"
    
# Derived Class 2
class Cat(Animal):
    def __init(self): pass
    def speak(self):        # overriding speak() function of parent class
        return "Meow !"
    
dog = Dog()
cat = Cat()
print(cat.speak())
print(dog.speak())
print()

# polymorphism using function
def makeSound(animal):
    print(animal.speak())

makeSound(cat)
makeSound(dog)
print()

dog = super(Dog, dog)       # typecasting
print(dog.speak())
```

```python
# Polymorphism using abstract class

# Abstract Base Classes (ABCs) are used to define common methods for a group of related class.
# Contains only function definition (no implementation)
# Syntax :- 
    # from abc import ABC, abstractmethod
    # class class_name(ABC):
    #     @abstractmethod
    #     def abstract_method1(): pass
    #     def abstract_method2(): pass

# Any abstract_method defined in the abstract class must be implemented in all its Child Class
```

```python
from abc import ABC, abstractmethod

# defining an abstract base class
class Vechicle(ABC):                # class is derived from ABC to make it abstract class
    @abstractmethod
    def start_engine(self):         # must be overridden by methods in child class
        pass
    
# derived Class 1
class Car(Vechicle):
    def __init__(self, doors, windows):
        self.doors = doors
        self.windows = windows
    
    def start_engine(self):         # must be overriden otherwise error
        return "Car Engine Started"
    
# derived class 2
class Motorcycle(Vechicle):
    def __init__(self, mileage):
        self.mileage = mileage

    def start_engine(self):         # must be overriden otherwise error
        return "Motorcycle Engine Started"
    
# function for polymorphism
def start_vechicle(vechicle):
    return vechicle.start_engine()

# creating objects
car = Car(4, 6)
motorcycle = Motorcycle(50)

print(start_vechicle(car))
print(start_vechicle(motorcycle))
```
