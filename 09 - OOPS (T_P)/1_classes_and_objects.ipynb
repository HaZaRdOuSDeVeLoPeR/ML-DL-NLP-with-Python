```python
# A class is a blueprint for creating objects. (Attributes and methods)

# Syntax :-
    # class <class_name>:           defining a class
    #     <constructor>     __init__(self)
    #     <instance_methods>

    # object = class_name()         creating an object

class sample:
    # constructor
    # self gives the reference of object on which the function is called (similar to "this" keyword in C++)
    def __init__(self, param1, param2):     
        self.param1 = param1
        self.param2 = param2
    
    def getInfo(self):
        return [self.param1, self.param2]

sample_obj = sample("param1", 20)
print(sample_obj.getInfo())

# Note :- 
# There is not need of declaring instance variables, we can use self keyword instead whenever we need to access/create the attributes
# There is no Access Modifier in Python. convention :- _var = protected    |    __var = private
```

```python
# Modelling a Bank Account

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{self.owner} deposited Amount of {amount}")
        print(f"New Balance is {self.balance}")

    def withdraw(self, amount):
        if(self.balance >= amount):
            self.balance -= amount
            print(f"{self.owner} withdrew Amount of {amount}")
            print(f"New Balance is {self.balance}")
        else:
            print("Insufficient funds")

    def getBalance(self):
        return self.balance
    
aditya_account = BankAccount("Aditya", 1000)
krish_account = BankAccount("Krish", 2000)

aditya_account.withdraw(2000)
aditya_account.deposit(500)

krish_account.deposit(1000)
krish_account.withdraw(2500)

print(krish_account.getBalance())
```
