```python
# magic methods are also known as dunder (double underscore) methods, which start and end with double underscore
# These methods enable to define the behaviour of objects for built-in operations such as arithmetic, comparisions etc
```

```python
# Some common magic methods are :-

#   __init__    :- initializes new instance of a class (constructor)
class Person:
    def __init__(self): pass

person = Person()
print(dir(person))  # list of functions defined for the object

#   __str__     :- returns string representation of an object
print(person)       # default str function is invoked

#   __repr__    :- returns "official" string representation of an object
#   __len__     :- returns the length of an object
#   __getitem__ :- gets an item from a container
#   __setitem__ :- sets an items in a container 
```

```python
# overriding __str__ method
class Person:
    def __init__(self, name, age):
        self.name = name    
        self.age = age 

    def __str__(self):
        return f"Name: {self.name}\nAge: {self.age}"

    def __len__(self):          # overriding len() method
        return 2

person = Person("Aditya", 22)
print(person)                   # Here, overriden magic method __str__() is called

print(len(person))              # calling overidden len() method
```
