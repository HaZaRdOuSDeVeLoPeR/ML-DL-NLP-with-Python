```python
# Inheritance is a fundamental concept in OOP, allowing a class to inherit its attributes and methods from another class while having some of its own

# Syntax :- class ChildClass(Parent1, Parent2, Parent3, ...):
```

```python
# Single Inheritance (similarly define hierarchical and multi-level inheritance)
# Parent Class
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows        
        self.doors = doors    
        self.enginetype = enginetype

    def getInfo(self):
        return (f"Total windows: {self.windows}\nTotal doors: {self.doors}\nEngineType: {self.enginetype}")

    def drive(self):
        print(f"The person will drive the {self.enginetype} car")

# Child Class
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, selfDriving):
        Car.__init__(self, windows, doors, enginetype)
        self.selfDriving = selfDriving

    def getInfo(self):
        info = Car.getInfo(self)
        info += f"\nSelf Driving: {self.selfDriving}"
        return info
    
    def drive(self):
        print(f"Tesla supports self-driving: {self.selfDriving}")
        Car.drive(self)

car1 = Car(6, 5, "petrol")          # parent class object
car1.drive()
print(car1.getInfo())
print()

tesla1 = Tesla(4, 4, "electric", True)
tesla1.drive()
print(tesla1.getInfo())

# super() returns the instance of superclass object tied to current(self) object
# super(ChildClass, ChildObject) also does the same thing (more generic), can be used anywhere
```

```python
# Multiple Inheritance

# Parent1 Class
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Animal Class Method is called if not overridden by child class"

# Parent2 Class
class Pet:
    def __init__(self, owner):
        self.owner = owner

    def speak(self):
        return "Pet Class Method is called if not overridden by child class"

    def gotoOwner(self):
        return "Pet Class Method is called if not overridden by child class"

# Base Class
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    # def speak(self):
    #     return f"{self.name} says woof"
    
    def gotoOwner(self):
        return f"{self.name} goes to {self.owner}"
    
# creating object
dog = Dog("Buddy", "Aditya")
print(dog.speak())              # Animal Class Method will be called since it was inherited first
print(dog.gotoOwner())
```
