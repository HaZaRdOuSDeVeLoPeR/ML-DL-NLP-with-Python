```python
# Encapsulation involves bundling data and methods that operate on similar objects

# It is the concept of wrapping data and methods together as one unit
# It restricts direct access to some of the object's components
```

```python
# Access Modifiers in Python :- Python doesn't have private, protected and public access modifiers
# while defining an instance variable :-
# .variable   = public variable       | accessible from anywhere
# ._variable  = protected variable    | accessible inside class and its child class
# .__variable = private variable      | accessible only inside same class
```

```python
class Person:
    def __init__(self, pan, aadhar, name, age):
        self.__pan = pan            # private variable
        self._aadhar = aadhar       # protected variable
        self.name = name            # public variable
        self.age = age 

    def getPAN(self):
        return self.__pan

    def setPAN(self, pan):
        self.__pan = pan                                                             


class Student(Person):
    def __init__(self, pan, aadhar, name, age, university):
        Person.__init__(self, pan, aadhar, name, age)
        self.university = university

    def getPAN(self):           # private members not accessible by child class
        return self.__pan
    
    # to manipulate private memebers of parent class, use getters and setters
    def PANgetter(self):                    # getter 
        return Person.getPAN(self)
    def PANsetter(self, pan):               # setter
        Person.setPAN(self, pan)

    def getAadhar(self):        # protected memebers are accessible by child
        return self._aadhar
    
    def getInfo(self):
        return f"PAN: {Person.getPAN(self)}\nName: {self.name}\nAge: {self.age}\nUniversity: {self.university}"
    
    
student = Student("ABCD", 1234, "Aditya", 22, "NITW")
print(student.getInfo())
print(student.getAadhar())          # accessing protected variables using getter
print()

# print(student.getPAN())       --> will throw an access modifier error
print(student.PANgetter())          # accessing private variables outside the class using getter
student.PANsetter("WXYZ")           # modifying private variables outside the class using setter
print(student.getInfo())
```
