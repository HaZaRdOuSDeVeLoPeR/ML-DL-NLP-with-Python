```python
# all custom exception class should inherit "Exception" class
# custom exception is thrown using "raise" keyword
# syntax :-
#     class myexception(Exception):
#         pass
    
# # syntax :-
#     raise myexception

class dobexception(Exception):
    def message(self):
        print("Sorry, age is not within the range of 20-30")

try:
    age = int(input("Enter your age: "))
    if(age >=20 and age <=30):
        print("Allowed")
    else:
        raise dobexception
    
except dobexception as dobexcept:
    dobexcept.message()
```
