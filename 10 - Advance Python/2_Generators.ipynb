```python
# generators are a simpler way to create iterators.
# using "yield" keyword, it produces a series of values lazily(on the fly), without storing them in memory

# yield basically returns an object but continues the execution of function (if some loc is left)

def gen_squares(n):
    for i in range(1, n):
        yield i**2

generator_obj = gen_squares(6)
print(type(generator_obj))

try:
    while(True):
        # using next method to traverse through generator (iterator)
        print(next(generator_obj), end=" ")
except Exception:
    pass
```

```python
# practical uses :-

# 1. Reading Large Files - Read the file one line at a time instead of reading whole file and saving into memory

def read_large_file(file_path):
    file = open(file_path, 'r')
    for line in file:
        yield line.strip()

    file.close()

file_path = "large_file.txt"
for line in read_large_file(file_path):
    print(line)
```
