```python
# Iterators allow for efficient looping and memory management
# Provides a way to access elements of a collection without exposing the underlying structure

# can be created for any collection :- tuple/set/dict/list/str etc
```

```python
# create iterator object using "iter" function
# Syntax :- iterator = iter(collection)     --> returns the iterator object

mylist = [1, 2, 3, 4]
myiterator = iter(mylist)
print(type(myiterator))

# use next(iterator) function to get the elements sequencially
try:
    while(True):
        print(next(myiterator), end = " ")
except StopIteration as si:         # iterator keeps moving forward till it reaches end and raises an exception
    print(si)
```
