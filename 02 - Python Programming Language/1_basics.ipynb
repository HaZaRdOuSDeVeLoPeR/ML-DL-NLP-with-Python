```python
# basic syntax and semantic rules in python
# using function defined in cython code cell
# print("Using Cython : ",add(5,6))

# 1) python is case sensitive

name = "Aditya"
Name = "Vimal"
print(name + " " + Name)        # by default python adds newline
print('\n', end="")             # to change this behaviour define own endline character using "end" argument of print function    

####################################################################

# 2) Python uses indentation to define a block of code. 
#    Other programming languages like C++ use symbols like {}
#    Enforces consistent and readable code

if True:
    print("Correct Indentation")
    if False:
        print("This will not Print")
    print("This will Print")
print("Outside if Block")
print('\n', end="")

####################################################################

# 3) Continue a single line on next line using '\' operator
#    Write multiple lines in single line using ';' operator

total = 2 + 3 + 4 + \
5 + 6
print("Single Line in Multiple Lines", total)

a = 5; b = 6; c = 7
print("Multiple Lines in Single Line", a + b + c)
print('\n', end="")

####################################################################

# 4) Python is a loosely typed (dynamically typed language)

x = 15
int_string = 18
print(x + int_string)                 # here type is int

myname = "Aditya"
int_string = "18"
print(myname + " " + int_string)        # here type is string
print('\n', end="")
```
