```python
# declaring and assigning variables

age = 22
height = 5.9
name = "Aditya Vimal"
isStudent = True

## print the variables and its types

print("Age :", age, "| Type :",type(age))
print("Height :", height, "| Type :",type(height))
print("Name :", name, "| Type :",type(name))
print("Is_Student :", isStudent, "| Type :", type(isStudent))
print('\n', end="")
```

```python
# Python is dynamically typed language. Type is determined at runtime
# In above example :-
#   age : int           | int
#   height : float      | float
#   name : string       | str
#   isStudent : boolean | bool

# Note :- all the primitive datatypes are immutable (non-modifiable)
```

```python
# typecasting is possible by specifying the datatype()

age_str = str(age)              # converting int to string
print("Age :", age_str, "| Type :",type(age_str))

string = "23"                   # converting string to int
string_int = int(string)
print("String_To_Int :", string_int, "| Type :",type(string_int))

height_int = int(height)        # converting float to int
print("Height :", height_int, "| Type :",type(height_int))
print('\n', end="")
```

```python
# dynamic typing is possible
# variable type can change as execution goes

dynamic_typed_var = 23
print("Initial Type :", type(dynamic_typed_var))
dynamic_typed_var = "Aditya"
print("Final Type :", type(dynamic_typed_var))
```

```python
# taking input from user using "input()" function
# syntax :- variable = input("Message")

# by default input function takes input as string datatype
user_input = input("Enter the age :")       
print(type(user_input))

# we need to explicity typecast if we need another datatype
user_input = int(input("Enter your age :"))
print(type(user_input))
```
