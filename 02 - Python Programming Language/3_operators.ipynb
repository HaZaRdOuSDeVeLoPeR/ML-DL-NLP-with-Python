```python
# arithmetic operators
a = 10
b = 4

# usual operators
print("Addition :", a + b)
print("Subtraction :", a - b)
print("Multiplication :", a * b)
print("Normal Division", a / b)         # exact division (float)
print("Modulus Operation :", b % a)
print('\n', end="")

# additional operators
print("Floor Division :", a // b)       # integer division
print("Exponentiation Operator :", b ** a)
```

```python
# comparision Operators
str1 = "Aditya"
str2 = "Vimal"

print(str1 == str2)
print(str1 != str2)

print(str1 <= str2)         # comparision in lexographic order
print(str1 < str2)
print(str1 >= str2)
print(str1 > str2)
```

```python
# logical operators 
# AND, OR, NOT

X = True
Y = False

print("T && F :", X and Y)
print("T && !(F) :", X and (not Y))
print("T || F :", X or Y)
print("!F :", not X)
```

```python
# bitwise operators
# [AND &] [OR |] [Complement ~] [XOR ^] [Left Shift <<] [Right Shift >>]

a = 5      # 0b0101
b = 3      # 0b0011

print(a & b)   # 1  (0b0001)
print(a | b)   # 7  (0b0111)
print(a ^ b)   # 6  (0b0110)
print(~a)      # -6 (two's complement)
print(a << 2)  # 20 (0b10100)
print(a >> 1)  # 2  (0b0010)
```
