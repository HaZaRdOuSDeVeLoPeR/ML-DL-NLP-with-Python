```python
# python modules and packages helps organize and reuse code.
# Syntax to import a package :- import <package_name>
# Syntax to import a specific function/class of a package :-
#       from <package_name> import function1, function2, ...
#       from <package_name> import *        --> import everything
```

```python
import math
print(math.sqrt(24))
print(math.pi)
```

```python
from math import sqrt, pi
print(sqrt(36))
print(pi)
```

```python
from math import *
print(sin(pi/2))
print(log(2)/log(e))
```
