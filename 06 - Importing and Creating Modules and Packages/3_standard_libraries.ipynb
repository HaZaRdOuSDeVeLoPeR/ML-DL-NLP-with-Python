```python
# data serialization

student = {                                 # data in form of json
    'name': 'Aditya Vimal',
    'education': {
        'degree': 'B.Tech',
        'branch': 'CSE',
        'roll_no': '23CSB0B30'
    },
    'age': 30,
    'blood_group': 'O+'
}

import json

stringified = json.dumps(student)
print(stringified)
print(type(stringified))

parsed = json.loads(stringified)
print(parsed)
print(type(parsed))
```

```python
# working with csv file
import csv

# writing to a csv file
file = open('example.csv', 'w', newline='')
writer = csv.writer(file)
writer.writerow(['name', 'age'])
writer.writerow(['Krish', 32])
writer.writerow(['Aditya', 22])
file.close()

# reading from a csv file
file = open('example.csv', 'r')
reader = csv.reader(file)
for row in reader:
    print(row)
file.close()
```

```python
# date operations

from datetime import *

today = datetime.now()
print(today)

yesterday = today - timedelta(days = 1)
print(yesterday)
```
