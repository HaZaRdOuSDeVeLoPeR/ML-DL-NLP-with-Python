```python
# A cost function is a function which measures how wrong a model predictions are compared to actual values
# It is denoted by :-
#                   J(y, yᵖ)      where, y is the actual value
#                                        yᵖ is the predicted value 
# It taken in the predicted and actual values as input and gives the total error in the model as output

# Various cost functions used are :-
#
# 1. Mean Square Error     :-  J = (1/n) * ∑[(yᵢ - yᵢᵖ)²]                          
#       - Smooth and Differentiable everywhere. 
#       - Good for Gradient Descent (Has One Global and Local Minima)
#       - Disadvantage :
#               - Prone to Noise/Outliers (penalizes large errors heavily)
# 
# 2. Mean Absolute Error    :-  J = (1/n) * ∑[|yᵢ - yᵢᵖ|]
#       - Robust to Outliers
#       - Disadvantage : 
#               - Not Differentiable at Zero. 
#               - Tricker to use for Gradient Descent
#
# 3. Root Mean Square Error :-  J = sqrt(∑(yᵢ - yᵢᵖ)²/n) = sqrt(Mean Square Error)
#       - Same Advantage/Disadvantage as MSE
#       - Converges slower than MSE
# 
# 4. Log Loss Function      :- J = (-1/n) * ∑[yᵢ.log(yᵢᵖ) + (1-yᵢ).log(1-yᵢᵖ)]
#       - Used in Logistic Regression, Neural Networks and Binary Classification
# 
# 5. Cross Entropy Function :- J = - ∑[yᵢ.log(yᵢᵖ)]
#       - Used in Multi-Class Classification Algorithms
# 
# 6. Hingle Loss Function   :- J = max(0, 1 - y.ȳ)
#       - Used in SVM
```
