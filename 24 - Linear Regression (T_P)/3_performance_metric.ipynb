```python
# Performance/Accuracy of the Regression Model is measured using two metrics :-

# 1. R (Coefficient of Determination)
# 
#       - It determines how much variation in the dependent variable is explained by trained regression model.
#       - Lies between [0, 1] ==> [Worst Fit, Perfect Fit]
#       - R >= 0.7 is considered Good (Explains 70% variability)
# 
#       - Given as :- R = 1 - (Sum_Sq_Residual/Sum_Sq_Totals)
# 
#                                     where, Sum_Sq_Residual = (yy)
#                                               Sum_Sq_Total = (yy)              
#  
#       - Disadvantage of Using R :-
#           - Lets say the model has u input parameters. The value of R is X
#             Lets say we introduce one more parameter which is not at all correlated with output feature
#             Still the value of R is going to increase.
#           
#           - Hence R is not considered a good technique for feature selection
#           - R always increases when you add more predictors, even if they are useless.


# 2. Adjusted R 
# 
#       - It fixes the problem of R (increasing when useless feature is added)
#       - Key idea is to penalize the useless predictors (features)
# 
#       - Given as :- 1 - (1 - R).(n - 1)/(n - p - 1)
#  
#                                       where, R is the coeff of determination
#                                              n = no of observations
#                                              p = no of predictors (features)


# Why Adjust R exists ?
# 
#   - Helps in selection of useful predictors (features) and discard useless ones
#   - Protects from Overfitting
#                   [Memoising Input Data instead of Training]
```
