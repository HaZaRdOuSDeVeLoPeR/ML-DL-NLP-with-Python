```python
# Multiple Linear Regression

# It is the generalization of Simple Linear Regression
# We have n input (independent) features and 1 output (dependent) feature

# Here we produce best fit Hyperplane instead of a line (in simple linear regression)

# Generalized Multiple Regression Formula
#       - Let y be the Output Feature
#       - Let [x₁, x₂, ..., xₙ] be n output features
#       - Let [β₀, β₁, β₂, ..., βₙ] be n + 1 model parameter [1 constant and n parameters corresponding to each input feature]
# 
#       - We can write :-
#                       y = β₀ + β₁x₁ + β₂x₂ + ... + βₙ₋₁xₙ₋₁ + βₙxₙ + ϵ 
```

```python
# Loss Function and Gradient
# 
# Loss Function for Multiple Regression can be given as :-
#               J = (1/n) * ∑(yᵢ - β₀ᵢ - β₁x₁ᵢ - β₂x₂ᵢ - ... - βₙxₙᵢ)²
# 
# On Computing Gradient w.r.t βᵤ we get :-
#                        ₙ    
#               ∂J/∂βᵢ = ∑ [βᵤ.(x̄ᵤx̄ᵢ)] - x̄ᵢȳ
#                       ᶸ⁼⁰  
#               where, x̄ᵤx̄ᵢ and x̄ᵢȳ are grouped means 
#                                                           At i = 0, use x₀ = 1
# 
# Note :- Put ∂J/∂βᵢ = 0 To get the Empirical formulas for each Parameters βᵢ ∀ i >= 0

# For Model Training, if parameters are standardized, original parameters in terms of normalized parameters can be expressed as :-
#       for i > 0 :         
#                   (βᵢ)⁽ᴼ⁾ = (σy⁽ᴼ⁾/σx⁽ᴼ⁾) * (βᵢ)⁽ⁿ⁾                     
#       for i = 0 :                                   ₙ 
#                   (β₀)⁽ᴼ⁾ = μy⁽ᴼ⁾ + σy⁽ᴼ⁾.(β₀)⁽ⁿ⁾ - ∑ (μxᵢ)⁽ᴼ⁾.(βᵢ)⁽ᴼ⁾ 
#                                                    ˣ⁼¹  
```
