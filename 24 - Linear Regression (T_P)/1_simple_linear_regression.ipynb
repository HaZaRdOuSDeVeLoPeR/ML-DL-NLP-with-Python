```python
# Simple Linear Regression

# It is a supervised ML algorithm.
# Predicts output(dependent) feature from a single input(independent) feature

# Eg :- Input Feature   ==> | Model | ==>     Output Feature
#           Weight      ==> | Model | ==>         Height
#            Age        ==> | Model | ==>         Weight

# Geometrical Interpretation :-
#
# In a scatter plot of (Input Feature, Output Feature), Find the best fit line
#       - Best Fit line is created by minimizing the perpendicular distance of each point from the line
#       - Minimize ∑(dᵢ) for each Pᵢ in the datapoint
# 
# Now this line is the generalized rule (Model) for the prediction of any new input feature

# Note :-
# In Simple Linear Regression, since there is single input feature and single output feature, we get a straight line as the Model
# In Multiple Regression, we get a hyperplane
```

```python
# Lets say we have prediction y w.r.t input x and the prediction has some error "ϵ"
# Then we can write the equation of regression model as :-
#                               y = β₀ + β₁.x + ϵ 
# 
# Lets say we have n input points [x₁, x₂, .., xₙ] and n prediction points [y₁, y₂, .., yₙ]
# as input for the training data.

# Then error in prediction for each input can be written as :-
#                               ϵᵢ = yᵢ - β₀ - β₁.x

# The error can be negative/positive so we need to minimize the magnitude of overall errors i.e
#                    = Minimize ∑|ϵᵢ| = Minimize |yᵢ - β₀ - β₁|
#                    = Minimize (yᵢ - β₀ - β₁.x)²
# 
# For Minimization, we take partial derivative w.r.t β₀ and β₁.
# We get :- 
#           β₀ = ȳ - β₁.x̄       and        β₁ = cov(x,y)/var(x)
#           where,
#                   x is the set of values for independent variable
#                   y is the set of values for dependent variable
```
