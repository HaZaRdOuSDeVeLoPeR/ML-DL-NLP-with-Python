```python
# There are two types of learning :-

# 1. Instance based Learning :- 
#       - It is also known as "lazy learning". No Training is Involved
#       - Main idea is to store the training data and make predictions for new data using already present data.
#       - It is memory heavy
#       - Eg :- KNN, Locally-Weighted Regression etc

# 2. Model based Learning :-
#       - It is also known as "eager learning". Actual Training is involved
#       - Main Idea is to build a Mathematical Model which is built on generalized pattern of training data
#       - It is memory efficient and predicts quickly. But training takes time.
#       - Eg :- Linear Regression, Decision Trees, Neural Networks, SVM etc
```
