```python
# Differences between AI, ML, DL and Data Science

# AI :- Goal is to develop an application which can perform its own task 
#       without any human intervention, mimicing human intelligence
#           Eg :-   Netflix Recommendation System
#                   Self Diving Car

# ML :- It is a Subset of AI. Provides stats tools, algorithms to 
#       "visualize", "analyze", "predict" and "forecasting" of data

# DL :- It is a subset of ML. Its goal is to make machine mimic human brain.
#       Mostly about Neural Network

# DS :- Involves all Three fields. In addition, some more mathematics

#       |-------------------------------------------|
#       | Artificial Intelligence                   |
#       |       |-------------------------------|   |
#       |       | Machine Learning       |------|---|------------|
#       |       |       |----------------|---|  |   |            |
#       |       |       | Deep Learning  |   |  |   |    Data    |
#       |       |       |----------------|---|  |   |   Science  |
#       |       |------------------------|------|   |            |
#       |--------------------------------|----------|            | 
#                                        |-----------------------|
```
