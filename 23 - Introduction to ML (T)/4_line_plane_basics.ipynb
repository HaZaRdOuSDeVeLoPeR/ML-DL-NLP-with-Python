```python
# Equation of Line, 3D-Plane and Hyperplane(n-dimensional plane)

# Equation of Straight Line :-  y = m.x + c     
#       - here, m ==> movement in y-axis with a unit increase in x-axis
#       - here, c ==> value of y when x = 0


# For a linear equation of n-dimensions (Hyperplane). We have multiple dependent variables
# 
#       - Generalized Eq :- w₁.x₁ + w₂.x₂ + w₃.x₃ + .... + wn.xn + b = 0
#
#       - We can represent it in matrix form :- WᵀX + b = 0
#           - where, W = [w₁, w₂, w₃, .., wₙ] ==> Vector of coefficients
#           - where, X = [x₁, x₂, x₃, .., xₙ] ==> Vector of variables
# 
#       - Note :- If the Hyperplane passes through (x₁, x₂, x₃, .., xₙ) = (0, 0, 0, .., 0)
#                   - Then Equation Becomes WᵀX = 0     ==>    i.e b = 0


# Mathematical Interpretation of generalized hyperplane equation passing through Origin
# 
#       - For WᵀX = 0, if we consider W and X to be n dimensional vectors
#       - We write :-
#                   W.X = 0     ==>     |W|*|X|*cosθ = 0
#
#       - Neither |W| or |X| can be 0. Hence "cosθ must be 0" or "θ = 90ᴼ" 
# 
#       - Conclusion :- 1. X is a set of points on Hyperplane.
#                       2. W is a vector perpendicular to that hyperplane.
#                       3. Hyperplane passes through origin
```
