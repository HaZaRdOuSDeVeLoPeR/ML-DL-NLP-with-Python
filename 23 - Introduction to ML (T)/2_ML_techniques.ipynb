```python
# Types of Machine Learning Techniques :-

# ----------------------------------------------------------------------------------------

# 1. Supervised Machine Learning Technnique
#       - Dataset given will have two kinds of features :-
#           - independent (input) feature
#           - dependent (output) feature
 
#       - Goal is to make a Model M such that (Output Feature) = M(Input Features)
#                                               [with highest possible accuracy]

#       - Here we basically solve two kinds of problems :-
#           - i) Regression Problem
#                   - Here Output Feature is Continuous or Probabilistic
#                   - Input Feature can be both Continuous and Categorical
#                   - Eg :- Linear, Ridge&Lasso, ElasticNet
# 
#           - ii) Classification Problem
#                   - Here Output Feature is Categorical (Discrete)
#                   - Input Feature can be both Continuous and Categorical
#                   - Eg :- Logistic Regression, Naive Bayes, SVM, Decision Tree, KNN etc.

#           - Note :- Decision Tree, Random Forest, AdaBoost, XgBoost 
#                     can solve both kinds of problems 

# ----------------------------------------------------------------------------------------

# 2. Unsupervised Machine Learning Technique
#       - Dataset doesn't constitute of any features (labels)
#       - Goal is to find similar patterns in the given data

#       - Various methods used are :-
#           - i) Clustering
#                   - Here, we group similar datapoints together to form clusters
#                   - Eg :- K-Means, Hierarchical, DBSCAN etc.
# 
#           - ii) Association Rule Learning
#                   - Here, we try to find relationships between items that appear together
#                   - Eg :- Market Basket Analysis [Apriori, DIC, FP-Growth etc], Recommendation Systems etc.

#       - Other problems solved using this technique are :-
#                   - Dimensionality Reduction      - Density Estimation
#                   - Anomaly Detection             - Feature Learning

# ----------------------------------------------------------------------------------------

# 3. Reinforcement Learning Technique
#       - Here the Agent (on which ML Model is deployed) learns by 
#         choosing its own actions and observing the results, receiving rewards
#         and improving itself through trial and error

#       - Goal of the Agent is to learn a Model which "maximizes total rewards" over time
```
