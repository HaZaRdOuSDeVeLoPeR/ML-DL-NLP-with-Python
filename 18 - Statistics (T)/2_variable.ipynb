```python
# Variable is an object which can take up any values

# Types of Variables
#   |
#   |------> Quantitative Variable
#   |         |
#   |         |-------> Discrete Quantitative 
#   |         |         (Values taken by variable are defined at regular intervals)  
#   |         |                                             Eg:- age, money
#   |         |-------> Continuous Quantitative
#   |                   (Can take infinite no of values)    Eg:- height, weight
#   |
#   |
#   |------> Qualitative/Categorical Variable
#               (A variable which can have only finite set of values)
#               Eg :- Gender (Male, Female)    Condition (Old, New)
```
