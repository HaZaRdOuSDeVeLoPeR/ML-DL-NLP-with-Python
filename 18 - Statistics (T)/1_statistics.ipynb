```python
# Two Main Types of Statistics :-

# 1. Descriptive Statistics :-
#  
# It consists of "organizing" and "summarizing" the data. Two main techniques are :-
#           i) Measure of central tendency (mean, mode, median)
#          ii) Measure of dispersion (variance and standard deviation)


# 2. Inferential Statistics :-
# 
# From a population data (bulk data) 
# we sample some data (filter out some data) using various sampling methods and 
# try to draw some conclusion/inferences (probilistic)
# by performing various experiements on the sampled data (Z-test, T-test etc)
```

```python
# Variance in terms of population data and sample data is calculated differently

# Q Why Sample variance is divided by "n-1" instead of "n" incase of population variance ?
# 
# - We do this to correct for bias introduced due to sampling (sampled data is only an estimate of true population)
# - This is also known as "Bessel's correction"

# Taking sqrt(variance), we get standard deviation.
#       - Variance tells about how much the data is dispersed all together
#       - Standard deviation tells about how much each data point is away from the mean on average
```
