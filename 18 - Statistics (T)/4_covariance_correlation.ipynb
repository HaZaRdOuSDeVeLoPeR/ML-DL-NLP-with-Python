```python
# Covariance and Correlation

# Covariance and Correlation are two statistical measures used to determine how much two random variables are dependendent on each other
# In other words, they quantify the degree of relationship between two random variables

# When X and Y are directly proportional, covariance and correlation is +ve
# When X and Y are inversely proportional, covariance and correlation is -ve
# When X and Y have no relation at all (random), covariance and correlation is 0
```

```python
from IPython.display import Image, display
display(Image(filename = "./assets/covariance.png"))
```

```python
# covariance of X w.r.t X (itself) is nothing but the Variance
# We can expand the idea and define the covariance of X, Y

# Advantage of using Co-Variance :-
#   - Helps in quantifying the relationship between two variables X and Y

# Disadvantage of using Co-Variance :-
#   - The range of covariance is not fixed (range is -inf to +inf)
#   - Degree of relationship can't be compared using the magnitude of covariance
#   - To overcome this problem, we normalize the value of covariance and get "correlation"
```

```python
from IPython.display import Image, display
display(Image(filename = "./assets/correlation.png"))
```

```python
# Correlation

# It is the normalized version of co-variance (range is fixed)
# Types of Correlation
#   |
#   |------> "pearson correlation coefficient" (range between -1 and +1)
#   |           - obtained by divided the cov by standard deviation of X and Y
#   |           - can only capture linear data-points perfectly
#   |           - fails to perfectly capture the non-linear correlation
#   |
#   |------> "spearman rank correlation"
#   |           - Obtained by computing Rank(X), Rank(Y) then computing its pearson coefficient
#   |           - can capture both linear/non-linear correlation perfectly

import numpy as np
from matplotlib import pyplot as plotter

# both variables are perfectly +vely correlated
X = np.array([x for x in range(1, 20)])
Y = np.array([x**4 for x in X])
plotter.plot(X.tolist(), Y.tolist(), marker = '*')

# pearson-coefficient fails to measure perfect non-linear correlation
print("Pearson-correlation-coeff:", np.corrcoef(X,Y)[0][1])

Rank_X = X.argsort()
Rank_Y = Y.argsort()

# spearman measures all kinds of correlation perfectly
print("Spearman-correlation coefficient:", np.corrcoef(Rank_X,Rank_Y)[0][1])
```
