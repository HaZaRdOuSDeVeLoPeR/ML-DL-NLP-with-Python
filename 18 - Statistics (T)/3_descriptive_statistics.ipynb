```python
# Histogram

# A technique to categorize the data into discrete bins of fixed width.
# Basically a bar-chart with categories as bins and frequency as values
# It also helps in estimating probability density function using "Kernel Density Estimation"

# The shape of probability density function is nothing but the histogram with bin size tending to 0
```

```python
# Percentile and Quartiles

# Percentile is a value below which a certain percentage of observations lie.
# Percentile(X) = [No. of observations below(X) / Sample Size]*100

# Eg :- If a value X has a percentile Y, then we can say that :-
#       --> Y% of observations in the sample lies below the value X

# Quartiles is dividing the data into 4 percentile slabs
# [0-25]    :- First Quartile
# [25-50]   :- Second Quartile
# [50-75]   :- Third Quartile
# [75-100]  :- Fourth Quartile
```

```python
# 5 Number Summary of Data are :-

# Minimum           - Minimum Observation in the sample
# 1st Quartile      - Value below which (25%) of observations lie
# Median            - Median of data (central tendency)
# 3rd Quartile      - Value below which (75%) of observations lie
# Maximum           - Maximum Observation in the sample
```

```python
# Outlier Detection :- 

# To identify outliers in the data, we need to identify the "lower fence" and "higher fence"
# Anything below lower fence and above higher fence is an outlier

# Lower Fence = Q1 - 1.5 (Inter Quantile Range)         where,
# Higher Fence = Q3 + 1.5 (Inter Quantile Range)        Inter Quantile Range = Q3 - Q1

import pandas as pd
arr = pd.Series([1,2,2,2,3,3,4,5,5,5,6,6,6,6,7,8,8,9,27])
print("Original Data:", arr.tolist())

Q1 = arr.quantile(0.25)     # compute 1st and 3rd quantiles
Q3 = arr.quantile(0.75) 
IQR = Q3 - Q1               # inter-quantile range

filtered_idx = (arr > Q1 - 1.5*IQR) & (arr < Q3 + 1.5*IQR)      # filter the outlier indices
arr = arr[filtered_idx]                 
print("Removed Outliers:", arr.tolist())
```
