```python
# WineQuality-Red Dataset
# Input feature     : All other features except Quality
# Output feature as : Quality

import pandas as pd

df = pd.read_csv('./assets/winequality-red.csv')
df.head(5)
```

```python
# List the dataframe 
# - columns_name and their their datatypes
# - no of rows
# - no of missing values in each column
# 
# using .info() method

df.info()
```

```python
# Generate the 5-Number Summary (and additional descriptive stats) for each column
# 
# using .describe() method

df.describe()
```

```python
# get all the unique values in a dataframe using .unique() method
print("Qualities of Wines:", df['quality'].unique())
```

```python
# Check for Missing Values in each column using .isnull() method
# Check for Non-missing values in each column using .notnull() method

# Both methods return a dataframe of same shape. 
# For each cell value, If there is a missing value then True otherwise False
df.isnull()
```

```python
# apply reduction accross rows (axis = 0) or columns (axis = 1)
# using .any() - checks whether atleast one value along the axis is true
# using .all() - checks whether all the values along the axis is true
# using sum(), mean(), std() etc

# computing total missing values in each row
df.isnull().sum(axis = 0)   # Sums up all True values in each column (reducing along row)
```

```python
# get df with Trues then group along columns to get rows without missing values
rows_without_missing_values = df.notnull().all(axis = 1)

df[rows_without_missing_values].head()      # removing the rows with missing values
```

```python
# checking for duplicate rows using .duplicated() method

duplicate_rows = df.duplicated()
unique_rows = duplicate_rows.apply(lambda x: not x)
print(unique_rows)

# drop duplicates using .drop_duplicates() method
df.drop_duplicates(inplace=True)
df.shape
```

```python
# Analysising the correlation of each feature with the other
corr_matrix = df.corr()

from matplotlib import pyplot as plotter
import seaborn as sns
plotter.figure(figsize=(12,12))
sns.heatmap(corr_matrix.abs(), annot=True, vmin=0, vmax=1)

corr_matrix.head(100)
```

```python
# Visualization
# count the frequency of each quality wine

freq_df = df.groupby(['quality'])[['pH']].agg('count')
freq_df.rename({'pH': 'count'}, inplace = True, axis = 1)
freq_df.reset_index(inplace = True)
print(freq_df)

plotter.xlabel('Wine Quality')
plotter.ylabel('Count')
plotter.bar(freq_df.iloc[:,0], freq_df.iloc[:,1])
```

```python
import math
import numpy as np
import matplotlib.pyplot as plotter

prob_df = df.copy()
normalized_df = df.copy()

def normal_pdf(x, mean, std):
    return (1/(std * math.sqrt(2*math.pi))) * math.exp(-((x - mean)**2)/(2*std**2))

for column in df.columns:
    if column == 'quality':
        continue
    
    # compute z-score normalized column
    mean = df[column].mean()
    std = df[column].std()
    if(std < 0.1): continue
    normalized_df[column] = df[column].apply(lambda x: (x - mean)/std)
    
    # compute PDF values
    prob_df[column] = df[column].apply(lambda x: normal_pdf(x, mean, std))

    # sort values so curve is correct
    sorted_idx = normalized_df[column].argsort()
    x_vals = normalized_df[column].iloc[sorted_idx]
    y_vals = prob_df[column].iloc[sorted_idx]
    
    plotter.plot(x_vals, y_vals, label=column)

plotter.legend()
plotter.show()
```
