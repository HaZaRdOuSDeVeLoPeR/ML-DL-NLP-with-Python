```python
# example 1

def palindrome_check(string):
    processed = [c.lower() for c in string if c.isalpha()]
    print(processed)
    return processed == processed[-1::-1]

print(palindrome_check("A man a plan a canal Panama"))
```

```python
# example 2

def count_words(file_path):
    word_freq = {}
    file = open(file_path, 'r')

    for line in file:
        words = line.split()
        for word in words:
            if word in word_freq:
                word_freq[word] += 1
            else: word_freq[word] = 1
    
    print(word_freq)

count_words("sample.txt")
```
