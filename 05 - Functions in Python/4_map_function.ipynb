```python
# map function applies a given function to all items in a collection (iterable)
# and returns a map object (an iterator).
# Syntax :-
#       map(function, iterable)     --> returns an iterator to map object

numbers = [1, 2, 3, 4, 5, 6, 7, 8]
map_object = map(lambda num: num**2, numbers)       # can also give normal function as argument
print(map_object)
print(list(map_object))         # typecasting the iterator to list


# example
people=[
    {"name":"Aditya","age":22},
    {"name":"Vimal","age":23}
]

map_object = map(lambda person: person["name"], people)
```

```python
# mapping muliple iterables
# syntax :- map(function, iterable1, iterable2, iterable 3, ....)

num1 = [2, 5, 6, 8]
num2 = [4, 6, 1, 9]
num3 = [1, 5, 7, 5]

map_object = map(lambda *args: sum(args), num1, num2, num3)

print(map_object)
print(list(map_object))
```
