```python
# lambda (anonymous) functions are simply a functions without a name
# Syntax :-
#   lambda argument(s): expression

# Note :- In python, lambda functions :-
#         - can't contain multiple statements
#         - can't contain return values
#         - can't have loops inside
#         - can't define any variables inside


add = lambda a,b: a+b   # lambda function
print(type(add))
print(add(3,5))
```
