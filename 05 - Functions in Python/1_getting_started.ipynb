```python
# functions is a block of code that performs specific task
# it helps in organizing code, reusing code and improving readibility

#   Syntax :-
#   def function_name(parameters):
#       function body
#       return value

def oddEvenCheck(a):
    return f"{a} is Even Number" if a % 2 == 0 else f"{a} is Odd Number"

print(oddEvenCheck(int(input("Enter a number: "))))
```

```python
# default parameter in function (similar to C++)

def greet(name = "Guest"):
    print(f"Hello {name}")

greet("Krish")
greet()                 # takes default value if parameter is missing
```

```python
# variable length arguments in pythonic function

# positional arguments using splat (*) operator
def pos_example(*args):
    for arg in args:
        print(arg, end = " ")
    print()

pos_example(1, 2, 3, 4.7, True, "Aditya")

# keyword arguments using splat (**) operator (catches all key_value pair arguments)
def kw_example(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}", end = " ")
    print()

kw_example(name = "Aditya", age = 22, college = "NIT Warangal")

# if a function has both keyword and positional arguments then, 
# all positional arguments must be defined before keyword arguments
```
