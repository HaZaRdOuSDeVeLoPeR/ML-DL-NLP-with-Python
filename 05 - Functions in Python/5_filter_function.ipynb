```python
# filter out items from a collection based on a given condition
# Syntax :- filter(bool function, iterable)
#   iterates through elements and filters the items which returns true
#   returns the iterator to filter object

my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]

def even_filter(num):
    return num % 2 == 0

def odd_filter(num):
    return num % 2 !=0

filter_object = filter(even_filter, my_list)
print(filter_object)
print(tuple(filter_object))

filter_object = filter(odd_filter, my_list)
print(filter_object)
print(tuple(filter_object))
```
