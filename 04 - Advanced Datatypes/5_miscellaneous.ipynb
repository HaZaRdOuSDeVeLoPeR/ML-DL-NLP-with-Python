```python
# f-string to get complete control over print statements
#   Syntax print(f"message{object}message ...")

my_list = ["Aditya Vimal", "Jeevesh Mittal", "Uddhav Nagar", "Manish Kumar"]
for name in my_list:
    print(f"/|__ {name} /|__ {id(name)}")
```

```python
# making copies in python
my_obj = {1:["Aditya", "Vimal"], "fruits":["mango", ("guava", "lichi")], "rank":5}

# using = operator always creates reference objects (same object diff name)

reference = my_obj
reference[1][1]=("Kumar")       # both are modified
print(my_obj)
print(reference)
print()


# using .copy() creates a shallow copy [copy of outer level, inner levels are still shared]

shallow_copy = my_obj.copy()
shallow_copy["rank"] = 1        # only shallow copy is modified
shallow_copy[1][1] = "Vimal"    # both are modified
print(my_obj)
print(shallow_copy)
print()


# using copy.deepcopy() creates a complete recursive copy of the object
import copy
deepcopy = copy.deepcopy(my_obj)

deepcopy[1] = ("Raushan", "Kumar", "Singh")     # only deepcopy will get modified
deepcopy["rank"] = 3
print(my_obj)
print(deepcopy)
```

```python
# splat operators
# used in packing/unpacking iterables (dict, tuples, set and list)

#    * is used to pack/unpack the values
#   ** is used to pack/unpack the key_value pairs

list1 = [1, 2, 3, 4, 5]
list2 = [6, 7, 8]

# unpacking list/tuple
[a, *b, c] = list1
print(a)
print(b)
print(c)

# merging list/tuple
merged = [*list1, *list2]
print(merged)

dict1 = {1:"Aditya", 2:"Rohan", 3:"Jeevesh", 4:"Uddhav", 5:"Krrish"}
dict2 = {6:"Bishwajit", 7:"Harshit", 8:"Manish"}
print()

#####################################################################

# unpacking dict
[a, *b, c] = dict1.items()
print(a)
print(b)
print(c)

# merging dict
merged = {**dict1, **dict2}
print(merged)
```
