```python
# dictionary is an mutable, unordered collection of items
# stores data in (key:value) tuples [list of tuples]
# keys are unique and immutable

# dictionaries are created using {} operator
empty_dict = {}
print(type(empty_dict))
empty_dict = dict()
print(type(empty_dict))

# during declaration, keys and value are seperated with ':'
my_dict = {1:"Aditya", 2:"Vimal", 3:"Raushan", 3:"Ram"}     # duplicates ignored
print(my_dict)
```

```python
my_dict = {"name":"Aditya", "age":22, "student":True}

# accessing dictionary elements using [] operator. Use index as the key
print(my_dict["name"])
```

```python
# dictionary updatation (add/delete)
my_dict = {"name":"Aditya", "age":22, "student":True}

# 1. [] operator to add a new key
my_dict["blood_group"] = "O+"
print("Added New Key :", my_dict)

# 2. del keyword to delete a key-value pair
del my_dict["age"]
print("Removed age key :", my_dict)
```

```python
# dictionary methods
my_dict = {"name":"Aditya", "age":22, "student":True}

# 1. get(key, default) - accessing dictionary elements
print(my_dict.get("age"))
print(my_dict.get("blood_group", "Not Available"))  # default value if key value is not present

# 2. keys() - method to get all the keys
print(my_dict.keys())

# 3. values() - method to get all the values
print(my_dict.values())

# 4. items() - get all key-value pairs
print(my_dict.items())

# 5. using in keyword to find keys/values
print("name" in my_dict.keys())
print(23 in my_dict.values())
```

```python
# iterating over dictionary
my_dict = {"name":"Aditya", "age":22, "student":True}

# iterating over keys using dict.keys()
for key in my_dict.keys():
    print(key, end=", ")
print()

# iterating over values using dict.values()
for value in my_dict.values():
    print(value, end=", ")
print()

# iterate over key_value pairs susing dict.items()
for (key, value) in my_dict.items():
    print(key, ":", value, end=", ")
print()
```

```python
# nested dictionary
students = {
    "student1":{"name":"Aditya", "age":22, "gender":"M"},
    "student2":{"name":"Vimal", "age":21, "gender":"M"}
}

# accessing nested dictionary
print(students["student2"]["gender"])       # access the gender of student2
```

```python
# dictionary comprehension
# similar to list comprehension, instead of one value, the expression will be key value pair

example = {x:x**2 for x in range(6)}
print(example)

conditional_example = {x:x**2 for x in range(10) if x % 2 == 0}
print(conditional_example)
```
