```python
# tuple is same as list. The only difference is, it is "immutable"
# tuples are ordered collection of immutable items.
# created using () instead of [] in list

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))


# tuple <---> list interconversion is possible by typecasting
# nested tuples, nested lists, nested tuples within list and vice versa is also possible

nested_tuple = ((1, 2), (3, 4, 5))
print(nested_tuple)
print(type(nested_tuple))

lists_within_tuple = ([1, 2], [3, 4, 5])
print(lists_within_tuple)
print(type(lists_within_tuple))
print(type(lists_within_tuple[0]))

tuples_within_list = [(1, 2), (3, 4, 5)]
print(tuples_within_list)
print(type(tuples_within_list))
print(type(tuples_within_list[0]))


# similar to lists, a tuple of mixed datatype is also possible
mixed_tuple = (1, 2.3, "Aditya", True)
print(mixed_tuple)
print([type(x) for x in mixed_tuple])


# accessing a element we will use [] operator similar to list
print(mixed_tuple[3])
```

```python
# tuple methods

# tuples are immutable
# hence only limited methods work

# slicing the tuple
mytuple = (0, 1, 2, 3, 4, 5)

# wrong way
# mytuple[3] = 5              

# correct way
# tuples of single element is written as (item, ) instead of (item)
newtuple = mytuple[:3] + (10, 10) + mytuple[4:]
print(newtuple)

# .count(item) - same as list (returns total occurences of item in tuple)
print("Count of 10 :", newtuple.count(10))
# .index() - same as list (returns the idx of first occurence of item)
print("First Occurence of 10 :", newtuple.index(10))

# append(), insert(), remove(), pop(), sort(), reverse() and clear() 
# are invalid in context to tuple (since they are immutable)
```

```python
# packing and unpacking a tuple

# packing
packed_tuple = 2, 3.5, ("Aditya", "Vimal"), False
print(packed_tuple)
print(type(packed_tuple))

# unpacking
(a, b, (c, d), e) = packed_tuple
print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
print(e, type(e))
```
