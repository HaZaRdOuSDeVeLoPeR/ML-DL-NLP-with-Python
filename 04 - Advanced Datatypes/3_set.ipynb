```python
# sets are used to store a collection of unique items
# doesnt allow duplicates
# doesnt preserve ordering (unordered set)
# sets are mutable

# set is created using {} operator
myset = {55, 111, 47, 33, 665, 33}
print(myset)
print(type(myset))

# typecasting between tuple <---> list <---> set is possible
mylist = [1, 2, 3, 4]
mytuple = (20, 30, 40, 40, 50)

typecasted = set(mylist)
print(typecasted, type(typecasted))
typecasted = set(mytuple)
print(typecasted, type(typecasted))
```

```python
# set methods
myset = {2, 44, 58, 98, 14}
print(myset)

# 1. add(item) - add an element to the set
myset.add(20)
print("Added 20 :", myset)

# 2. discard(item) - removes an element from set if present
myset.discard(44)
print("Discarded 44 :", myset)
myset.discard(11)     # not present hence does nothing
print("Discarded 11 :", myset)

# 3. clear() - clears the set (makes it empty)
```

```python
# set membership test
myset = {1, 2, 3, 4, 5}
print(4 in myset)           # test whether an item is present in set or not
print(10 in myset)


# set theory operations
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

# 1. union(set2) - takes union of set2 with the set on which method is called and returns it
union_set = set1.union(set2)
print(union_set)

# 2. intersection(set2) - takes intersection of set2 with set on which the method is called and returns it
intersection_set = set1.intersection(set2)
print(intersection_set)

# 3. difference(set2) - takes difference of set2 from set on which the method is called and returns it 
difference_set = set1.difference(set2)
print(difference_set)

# 4. symmetric_difference(set2) - removes common elements from both sets and returns unique elements to each set
symm_diff_set = set1.symmetric_difference(set2)
print(symm_diff_set)

# Note :- for each set theory operations, there is _update version for each method
#           union_update - modifies the set1 with union set
#           intersection_update - modifies the set1 with intersection_set
#           difference_update - modifies the set1 with difference_set
#           symmetric_difference_update - modifes the set1 with symmetric difference set

# 5. issubset() - verifies whether set1 is a subset of set2
print(myset.issubset(set1))

# 6. issuperset() - verifies whether set1 is a superset of set2
print(myset.issuperset(set1))
```
