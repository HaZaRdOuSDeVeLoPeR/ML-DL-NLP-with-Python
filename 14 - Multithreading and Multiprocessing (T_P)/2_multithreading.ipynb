```python
# When to use Multi-Threading ?

# Case 1:- If tasks are I/O bound (Tasks spend more time waiting for I/O operations)
# Case 2:- Concurrent Exection Required (Improves overall throughput by performing multiple operations concurrently)

# threading library in python provides the tools for creating multiple threads
```

```python
import threading
import time

# create a new thread
# Syntax :- t = threading.Thread(target)

# start the thread  :- t.start()
# join the thread   :- t.join()
```

```python
# simulating cpu heavy task
def cpu_heavy_task():
    for i in range(30):
        time.sleep(0.1)            # cpu_heavy task, sleeps for very less time
        print("CPU Heavy Task", end=" -> ")

# simulating i/o bounded task
def io_bound_task():
    for x in range(10):
        time.sleep(0.5)                # i/o bound task, cpu sleeps for more time     
        print("IO Bound Task", end=" -> ")
```

```python
# Case 1 :-  single threaded (sequential execution)

start = time.time()
io_bound_task()             # io_bound_task blocking the useful tasks
cpu_heavy_task()
end = time.time()

print()
print(f"Total Execution Time: {end-start}s")
```

```python
# Case 2 :-  multi threaded (parallel execution)

# create sepeate threads for two functions
cpu_heavy_thread = threading.Thread(target = cpu_heavy_task)
io_bound_thread = threading.Thread(target = io_bound_task)

# start the counter
start = time.time()

# start both threads
cpu_heavy_thread.start(); io_bound_thread.start()

# wait for the two threads to complete
cpu_heavy_thread.join(); io_bound_thread.join()

# stop the counter
end = time.time()

print()
print(f"Total Execution Time: {end-start}s")
```
