```python
# Program - Program is a sequence of instructions written in any programming langugage

# ----------------------------------------------------------------------------------------------------------

# Process - Process is an "instance of a program (from job queue to ready queue)" which is being executed

# Resources required while execution of a process :-
#   1. Code Segment - The program (instructions being executed itself)
#   2. Data Segment - Global variables
#   3. Heap Segment - Dynamically allocated memory
#   4. Stack Segment and Registers - Function calls, local variables, Call Stack

# Every Process has seperate memory space (hence once process can't corrupt the other)
#       --> due to this, it causes some overhead while context switching

# ----------------------------------------------------------------------------------------------------------

# Thread - It is a unit of execution within a process.
# A process is made up of process(es) - can be single-threaded or multi-threaded

# All threads within a process "share the same Code, Data and Heap Segment"
# All threads have "seperate stack segment and registers"
#       --> all threads share the same memory space (hence fast context switching)
```
