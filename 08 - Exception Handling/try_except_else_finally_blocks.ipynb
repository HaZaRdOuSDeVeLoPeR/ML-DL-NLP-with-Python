```python
# exception handling in python allows to handle errors gracefully 
# and take corrective measures without stopping the execution of the program

# exceptions are events that disrupts normal flow of a program
```

```python
# All exceptions are derived from a parent exception class "Exception"
# Some common exceptions are :-
    # NameError
    # FileNotFoundError
    # ZeroDivisionError
    # TypeError
    # ValueError

# Syntax :-
# try:
#     code_expected_to_raise_exception
# except <exception_class> as alias:
#     exception_handling_logic

try:
    a = 1/0
except ZeroDivisionError as divideby0:      # specific error class can catch
    print(type(divideby0))                  # errors of that class only
    print(divideby0)

try:
    a = b
except NameError as ne:
    print(type(ne))
    print(ne)

try:
    arr = [0, 1, 2, 3, 4, 5]
    print(arr[10])
except Exception as ex:    # parent Exception class can catch any exception
    print(type(ex))
    print(ex)
```

```python
# try .. except .. else block
# if none of the errors(exceptions) occur, only then else block is executed
import math

try:
    num = float(input("Enter a Number"))
    result = (math.sqrt(num))/num           # find 1/sqrt(input)
except ValueError as ve:
    print("Not a valid Number")
except ZeroDivisionError as zde:
    print("Division by Zero is Not Allowed")
except Exception as ex:
    print(ex)
else:
    print(f"Square Root is {result}")
```

```python
# try .. except .. else block
# finally block always gets executed regardless of exception raised or not
import math

try:
    num = float(input("Enter a Number"))
    result = (math.sqrt(num))/num           # find 1/sqrt(input)
except ValueError as ve:
    print("Not a valid Number")
except ZeroDivisionError as zde:
    print("Division by Zero is Not Allowed")
except Exception as ex:
    print(ex)
else:
    print(f"Square Root is {result}")
finally:
    print("Execution Completed")

# finally blocks are used to perform task(s) when a program finishes its execution
# such as saving files, terminating child processes etc
```
